```python
import Florian as flo
import gsw
wmo = 6900798
data = flo.read_data(wmo)
```

```python
# compute density
# see https://teos-10.github.io/GSW-Python/
psal = gsw.SA_from_SP(np.array(data['psal']), np.array(data['depth']), np.array(data['lon']), np.array(data['lat']))
temp = gsw.CT_from_t(psal, np.array(data['temp']), np.array(data['depth']))
sigma = gsw.sigma0(psal, temp)
data['sigma'] = sigma
```

```python
tmp = tmp[['depth', 'sigma']]
tmp.dropna(inplace=True)
#tmp.to_csv('prof_test2.csv', sep = ",")
```

```python
def mld_estimation_Huang(data):
    
    # need to check the QCs but this step should be done before
    
    data = data[['depth', 'sigma']]
    data.dropna(inplace=True)
    
    depth = np.array(data['depth'])
    sigma = np.array(data['sigma'])
    
    sd_density = list()
    for i in range(len(sigma)):
        tmp = np.std(sigma[:i+1])
        sd_density.append(tmp)
    
    range_density = list()
    for i in range(len(sigma)):
        tmp = np.max(sigma[:i+1]) - np.min(sigma[:i+1])
        range_density.append(tmp)
        
    khi = np.array(sd_density)/np.array(range_density)
    
    zn1_index = np.concatenate(np.where(khi == np.nanmin(khi)))[0] # why did Louis put %>% max on a which.min
    if zn1_index == len(khi):
        zn1_index = zn1_index-1
        
    zn2 = list()
    for i in range(zn1_index):
        #print(i)
        zn2.append(float(np.diff(sd_density[i:i+2])/np.diff(sd_density[zn1_index:(zn1_index+2)])))
    
    zn2 = np.array(zn2) 
    
    if any(x<= 0.3 for x in zn2) == True:
        zn2_index = np.max(np.concatenate(np.where(zn2 <= 0.3)))
    else:
        zn2_index = zn1_index
        
    return(depth[zn2_index])
```
